```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('occlusion2.jpg')

# Define the kernel size
kernel_size = (5, 5)

# Define the kernel
kernel = np.ones(kernel_size, np.uint8)

# Apply closing operation
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

# Apply erosion
erosion = cv2.erode(closing, kernel, iterations=1)

# Calculate the boundary image
boundaryImg = closing - erosion

# Save the boundary image to a CSV file
boundaryImg_flat = boundaryImg.flatten()
boundaryImg_df = pd.DataFrame(boundaryImg_flat, columns=['Pixel_Value'])
boundaryImg_df.to_csv('boundaryImg.csv', index=False)

# Load the CSV file
boundaryImg_csv = pd.read_csv('boundaryImg.csv')

# Convert the CSV file back to a NumPy array
boundaryImg_np = np.array(boundaryImg_csv['Pixel_Value']).reshape(boundaryImg.shape)

# Use Matplotlib to display the boundary image
plt.imshow(boundaryImg_np, cmap='gray')
plt.title('Boundary Image')
plt.show()

# Wait for a key press
cv2.waitKey(0)

# Close all OpenCV windows
cv2.destroyAllWindows()
```
